```python
# Here, the foundations of Seaborn use are demonstrated: importing, setting a theme, loading data, and visualizing

# Import Seaborn as sns, which is the conventional alias
import seaborn as sns

# Apply the default theme, seaborn has a library of themes which alter the underlying matplotlib charts in preset ways...
# ...this can be more meticulously controlled with functions described later on
sns.set_theme()

# Load an example dataset; there are several example datasets pre-loaded into Seaborn. If creating visualizations or...
# ... analyses for a portfolio, it is good to stay away from common datasets, and focus on either industry-specific or...
# ... personally relevant datasets that can demonstrate more thought, interest, and industry-specific knowledge
tips = sns.load_dataset("tips")

# Creating the visualization; unlike in base Matplotlib, where variables such as color have to be declared in each rendering...
# ... of the visualization, Seaborn handles these in the background whilst asking for only the essential arguments of the ...
# ... visualization function -- data, the x and y axes, column, which variable is used for the hue and style, and size
sns.relplot(
    data=tips, 
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)

# replot() is named replot because it is deisgned to plot many different statistical relationships, hence, relationship-plot. Here...
# ... the default scatter-plot is used, and the style and size paramters autmatically affect the shape of the plotted points and...
# ... their size. If the graph were a linear one instead, the above parameters would autmoatically be affected differently.
# HENCE: Seaborn beats Matplotlib on efficiency and ease, but loses on detail and customizability.
```

```python
sns.relplot(
    data=tips, kind= "line",
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)

# Here, the parameter "kind" has been used. It allows you to designate the kind of visualization you want to be graphed. In this case...
# ... it damages the quality of the graph, as it is not suited to the x axis of cost. It is best suited, generally, for x axes of...
# ... time, as such, the previous graph ought to be used instead.
```

```python
# Statistical estimation of confidence interavls is another baked-in feature of Seaborn. Observe the following grpah:
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)

# Here, the average of the data is automatically taken and plotted, and the confidence interval of said average is likewise...
# ... automatically computed and represented graphically. As such, the nature of the data used is supremely important when...
# ... using Seaborn as a graphing instrument, as the form of the data can dictate the form of the graph.
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")

# Here, the very important regression line makes its debut. A simple scatterplot is given further usfulness through .lmplot()...
# ...which combines .regplot() (normal regression line) and a FacetGrid (scatterplot grid) to output the following graph:

# Regression models are relatively simple models that demonstrate the trend of a given dataset over a given variable (usually)...
# ... time. Although they hold some predictive power, their simplicity makes them poor choices for precise forecasting.
```
